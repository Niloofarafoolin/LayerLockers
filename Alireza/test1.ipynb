```python
import trimesh

# Load the STL file
mesh = trimesh.load("original.stl")

# Apply a rotation (e.g., 90 degrees about the Z-axis)
rotation_matrix = trimesh.transformations.rotation_matrix(angle=2.57, direction=[1, 0, 1])
mesh.apply_transform(rotation_matrix)

# Save the transformed STL file
mesh.export("rotated_object.stl")
```
